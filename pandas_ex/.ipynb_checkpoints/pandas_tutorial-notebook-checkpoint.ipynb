# Part 3: Important Data Formatting Methods
How to use merge, sort, reset_index, and fillna

### Pandas Merge (a.k.a. “joining” dataframes)

Merge 2 seperate tables together similar to SQL join

![title](https://data36.com/wp-content/uploads/2018/08/0-pandas-merge-768x587.png)

So they look like this 

![title](https://data36.com/wp-content/uploads/2018/08/1-pandas-merge-508x687.png)


```python
zoo_eats = pd.read_csv('zoo_eats.csv', delimiter=';'  
                           , names = ['animal', 'food'])  
```

```python
# Output zoo_eats
zoo_eats
```

```python
# Merge the 2 tables together
# NOTE: we are missing the lions
zoo.merge(zoo_eats)
```

### How Pandas does its merging

How SQL does its merge

when we do zoo.merge(zoo_eats) it was an INNER merge, to get lions back we want to do an outter merge so we specify that

```python
# Note doing this we are missing some data with giraffe regarding uniq_id
# and water_needed so we should try a left join next
zoo.merge(zoo_eats, how = 'outer')
```

```python
zoo.merge(zoo_eats, how = 'left')
```

### Pandas Merge. On which column?
For doing a merge pandas need the key-colum that you want to base the merge onto. In this case we cant it on the animals column on both tables. So we need to specify the merge on "left_on" or "right_on"

```python
# Note: again, in the previous examples pandas automatically found
# the key-columns anyway… but there are many cases when it doesn’t.
# So keep left_on and right_on in mind.
zoo.merge(zoo_eats, how = 'left', left_on = 'animal', right_on = 'animal')
```

### Sorting in pandas
Do it by using sort_values()
Specify parameters inside()


```python
zoo.sort_values('water_need')
```

```python
# Specifying with more than 1 parameter, it sorts in ascending order, but the
# values of each type are descending
zoo.sort_values(by = ['animal', 'water_need'])
```

```python
# Sorts by descending order now
zoo.sort_values(by = ['water_need'], ascending = False)
```

### Reset_Index
Reset the RHS index with reset_index()

```python
# NOTE: it also keeps the old index value as well
zoo.sort_values(by = ['water_need'], ascending = False).reset_index()
```

```python
# Remove the old index value
zoo.sort_values(by = ['water_need'], ascending = False).reset_index(drop = True)
```

### Fillna
Its basically fill+na 

```python
#Do a left-merge 
zoo.merge(zoo_eats, how = 'left')
```

```python
# By using fill-na we get rid of the NaN values with a certain parameter 
# since lions eat meat fill it with meat!
zoo.merge(zoo_eats, how = 'left').fillna('meat')
```

```python
blog_buy = pd.read_csv('pandas_tutorial_buy.csv', delimiter=';', names = ['my_date_time', 'event', 'user_id', 'amount'])
```

### Task 1
Find the mean revenue between 2018-01-01 and 2018-01-07 from the users in the article_read frame

```python
# first merge the tables together based on user_id
step_1 = article_read.merge(blog_buy,
                            how = 'left', left_on = 'user_id',
                            right_on = 'user_id')

# create a series based on amount
step_2 = step_1.amount

# fill the NaN with 0
step_3= step_2.fillna(0)

#find the mean of the series
result = step_3.mean()

#display the mean
result


# # Outputting in 1 line

# article_read.merge(blog_buy,
#                     how = 'left', left_on = 'user_id',
#                     right_on = 'user_id').amount.fillna(0).mean()
```

### Task 2
Print the top 3 countires by total revenue between 2018-01-01 and 2018-01-07 from the users in the article_read dataframe

```python
# first merge the 2 tables together based off user id
step_1 = article_read.merge(blog_buy,
                            how = 'left', left_on = 'user_id',
                            right_on = 'user_id')

# fill the empty data with 0
step_2 = step_1.fillna(0)

# group by countries and then sum amounts in each country
step_3 = step_2.groupby('country').sum()

# create a data frame of countries, amount
step_4 = step_3.amount

# Sort in descending order
step_5 = step_4.sort_values(ascending = False)

# List the top 3 countries
step_5.head(3)


# # 1 line solution
# article_read.merge(blog_buy,
#                             how = 'left', left_on = 'user_id',
#                             right_on = 'user_id').fillna(0).groupby(
#                             'country').sum().amount.sort_values(
#                             ascending = False).head(3)
```
